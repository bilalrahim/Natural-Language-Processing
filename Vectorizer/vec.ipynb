```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
corpus=open('C:\\Users\\Bilal\\Downloads\\Movies_TV.txt').read()
```

```python
res=re.sub(r'Domain.*\n','',corpus)
#Removing first line, which is unnecessary.
```

```python
data_list=res.split('\t')
```

```python
##Q2
```

```python
#Frequency
f_vect=CountVectorizer()
F=f_vect.fit_transform(data_list)
print(f_vect.vocabulary_)
```

```python
vect=CountVectorizer(binary=True)
```

```python
X=vect.fit_transform(data_list)
```

```python
X.toarray()
#In binary form
```

```python
##TfId
```

```python
vect2=TfidfTransformer()
```

```python
X.toarray()
```

```python
Y=vect2.fit_transform(X)
```

```python
Y
```

```python
Y.toarray()
```

```python
##Q3Keep ngrams..
```

```python
vect3=CountVectorizer(ngram_range=(1,1))
unigram=vect3.fit_transform(data_list)
unigram.toarray()
```

```python
vect4=CountVectorizer(ngram_range=(1,2))
bigram=vect4.fit_transform(data_list)
bigram.toarray()
```

```python
vect5=CountVectorizer(ngram_range=(1,3))
trigram=vect5.fit_transform(data_list)
trigram.toarray()
```
